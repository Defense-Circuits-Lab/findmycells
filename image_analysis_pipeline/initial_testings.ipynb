# Lets load our actual data

### Images:

```python
single_plane_images = os.listdir('/home/ds/Retune/3d_cell_counting/cropped_planes')
```

### Reconstruct a 3D array from the 2D predicted segmentation masks

```python
single_plane_segmentation_images = os.listdir('/home/ds/Retune/3d_cell_counting/cropped_masks/cell_segmentations')
single_plane_segmentation_images.sort()

single_plane_segmentation_arrays = []

for image in single_plane_segmentation_images:
    array = np.asarray(Image.open('/home/ds/Retune/3d_cell_counting/cropped_masks/cell_segmentations/{}'.format(image)))
    single_plane_segmentation_arrays.append(array)
    
stacked_3D_segmentation_array = np.asarray(single_plane_segmentation_arrays)
stacked_3D_segmentation_array.shape
```

### Load the segmentation of the substantia nigra

```python
sn_segmentation_array = np.asarray(Image.open('/home/ds/Retune/3d_cell_counting/cropped_masks/substantia_nigra_segmentations/0000_000_cropped.png'), 
                                   dtype=bool)
sn_segmentation_array.shape
```

### Turn all predicted pixels outside of the mask black

```python
masked_3D_stack = stacked_3D_segmentation_array.copy()

for plane in range(masked_3D_stack.shape[0]):
    masked_3D_stack[plane][~sn_segmentation_array] = 0
```

### Count connected components

```python
labels_out_3D_segmentation_array, N_3D_segmentation_array = cc3d.connected_components(stacked_3D_segmentation_array, return_N=True)
labels_out_masked_3D_stack, N_masked_3D_stack = cc3d.connected_components(masked_3D_stack, return_N=True)

print('We found {} connected components in the 3D segmentation array and {} in the masked 3D stack!'.format(str(N_3D_segmentation_array), str(N_masked_3D_stack)))
```

### Crop stack

Unfortunately, the entire stack is too huge to be plotted in 3D (at least using the matplotlib library & the `voxels()` method. So lets crop it into a small junk so that we can also visually inspect whether the count of connected components is actually correct

```python
cropped_stack = masked_3D_stack[:, 4250:4650, 1750:2150] #alternative slices: [:, 4250:4650, 1750:2150] or [:, 5000:5400, 2200:2500]

fig = plt.figure(figsize=(9,35), facecolor='white')
gs = fig.add_gridspec(9, 2)

for row in range(8):
    fig.add_subplot(gs[row, 0])
    img = np.asarray(Image.open('/home/ds/Retune/3d_cell_counting/cropped_planes/0000_00{}_cropped.png'.format(str(row + 1))))
    plt.imshow(img[4250:4650, 1750:2150])
    plt.ylabel('plane 00{}'.format(str(row + 1)))
    if row == 0:
        plt.title('images')

for row in range(8):
    fig.add_subplot(gs[row, 1])
    plt.imshow(cropped_stack[row])
    if row == 0:
        plt.title('masks')

plt.tight_layout()
#plt.savefig('plane_by_plane_cropped_area_02.png', dpi=300)
plt.show()
```

```python
labels_out_cropped_stack, N_cropped_stack = cc3d.connected_components(cropped_stack, return_N=True)
```

```python
fig = plt.figure(figsize=(10,10), facecolor='white')

ax3D = fig.add_subplot(projection='3d')
ax3D.voxels(cropped_stack)

plt.title('{} connected components were detected!'.format(str(N_cropped_stack)))
plt.savefig('3D_rendering_cropped_area_01.png', dpi=300)
plt.show()
```
